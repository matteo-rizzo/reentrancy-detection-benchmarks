```python

import os, csv
from pathlib import Path

# ----------------------------------------------------
# CONFIG: change this to the directory you want to scan
# ----------------------------------------------------
ROOT_DIR = Path("/path/to/your/solidity/project")
ROOT_DIR = Path("../benchmarks/RSD/src/")
#ROOT_DIR = Path("../benchmarks/aggregated-benchmark/raw/")
OUTPUT_CSV = Path("contracts_loc.csv")

def count_loc(path: Path) -> int:
    """Count non-empty lines of code in a .sol file."""
    loc = 0
    with path.open("r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            if line.strip():
                loc += 1
    return loc

# recursively find .sol files
sol_files = [
    Path(dirpath) / fname
    for dirpath, _, filenames in os.walk(ROOT_DIR)
    for fname in filenames
    if fname.lower().endswith(".sol")
]

print(f"Found {len(sol_files)} Solidity files.")

if OUTPUT_CSV.exists():
    with OUTPUT_CSV.open("a", newline="", encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
        
        for sol in sol_files:
            contract_name = sol.name                  # <-- KEEP .sol extension
            local_path = sol.relative_to(ROOT_DIR)    # relative path
            loc = count_loc(sol)
            
            writer.writerow([contract_name, str(local_path), loc])
else:
    with OUTPUT_CSV.open("w", newline="", encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["contract_name", "local_path", "loc"])
        
        for sol in sol_files:
            contract_name = sol.name                  # <-- KEEP .sol extension
            local_path = sol.relative_to(ROOT_DIR)    # relative path
            loc = count_loc(sol)
            
            writer.writerow([contract_name, str(local_path), loc])

print(f"CSV saved as {OUTPUT_CSV.resolve()}")

```
